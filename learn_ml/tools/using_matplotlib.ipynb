```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.plot(U, V)
plt.show()
```

```python
array_cols = [10, 20, 30, 40, 50]
array_rows = [1, 2, 3, 4, 5]
plt.xticks(array_cols)
plt.yticks(array_rows)

XX, YY = np.meshgrid(array_cols, array_rows)
plt.plot(XX, YY, marker=".", color='k', linestyle='none')

plt.show()
```

```python

def f(x):
    return 3 * x ** 2 - 4 * x

def df(x):
    return 6 * x - 4

def tf(x):
    return 2 * x - 3


x = np.arange(0, 3, 0.1)
dx = df(x)
tx = tf(x)


plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label="y = 3x^2 - 4x")
plt.plot(x, dx, label="d/dx f(x)")
plt.plot(x, tx, label="Tangent line (x=1)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = 3x^2 - 4x and tangent line at x=1")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the function
def f(x, y):
    return np.log(x**2 + y**2)

    #     return x**2 + y**2

# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of points
x, y = np.meshgrid(x, y)

# Calculate z values
z = f(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```

# Plotting various functions used in ML

```python
"""
Standard imports for plotting
"""
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.arange(-10, 11)

# Prepare the plot
fig,ax = plt.subplots(1,3,figsize=(15,6))

y = 2* x + 1
ax[0].plot(x,y)
ax[0].set_title("Linear function: y = 2x + 1")

y = x**2
ax[1].plot(x,y)
ax[1].set_title("Quadratic function: y = x^2")

# Exponential
y = np.exp(x)
ax[2].plot(x,y)
ax[2].set_title("Exponential function: y = e^x")


# Show the plot
plt.show()

fig,ax = plt.subplots(1,3,figsize=(15,6))
for a in ax:
    a.grid(True)


x = np.linspace(0.0001, 20)
# Logarithmic
y = np.log(x)
ax[0].plot(x,y)
ax[0].set_title("Logarithmic function: y = log(x)")

# Logarithmic
y = -np.log(x)
ax[1].plot(x,y)
ax[1].plot(1, -np.log(1), marker="o", color="red")   
ax[1].set_title("Logarithmic function: y = -log(x)")


# Sigmoid
x = np.linspace(-10, 10)
y = 1 / (1 + np.exp(-x))
ax[2].plot(x,y)
ax[2].set_title("Sigmoid function: y = 1 / (1 + e^(-x))")


# Show the plot
plt.show()
```

```python
x = np.arange(-3, 3)
y = np.exp(x)

plt.plot(x, y)
plt.plot(0, np.exp(0), marker="o", color="red")
plt.plot(1, np.exp(1), marker="o", color="red")
# Add grid lines to the plot
plt.grid(True)
plt.show()


```

Loss function of logistics regression is 

$$
L(y, \hat{y}) = \begin{cases}
 -\log(\hat{y}) & \text{if } y = 1 \\
 -\log(1-\hat{y}) & \text{if } y = 0
 \end{cases}
$$

```python
"""
Loss function
"""

y_hat = np.linspace(0.0001, 0.9999, 100)

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
# ax.plot(y_hat, np.log(y_hat), label="log(y_hat)")
ax.plot(y_hat, -np.log(y_hat), label="y = 1: -log(y_hat)")
ax.plot(y_hat, -np.log(1-y_hat), color="red", label="y = 0: -log(1-y_hat)")
ax.set_title("Loss functions for logistic regression")
ax.legend()
ax.grid(True)
plt.show()
```
