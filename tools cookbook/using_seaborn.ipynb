```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

# Intro

```python
df = pd.read_csv('../datasets/womens_clothing_ecommerce_reviews_transformed.csv')
df.head()
```

```python
sns.countplot(data=df, x='sentiment', hue='product_category')
plt.legend(loc='upper right',bbox_to_anchor=(1.3, 1.1))
```

## Built-in data

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.lmplot(x='pclass', y='survived', data=titanic_df, aspect=2, )
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
sns.set_style("whitegrid")
sns.lmplot(x='sepal_length', y='sepal_width', data=iris_df, aspect=2, )
```
