# How to use Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1D data

```python
# Prepare the data
x = np.linspace(1, 20)
y = np.cos(x)
z = np.sin(x)

# Plot it
plt.plot(y, z)
plt.show()
```

```python
x = np.linspace(0, np.pi*2, 100)
y = np.cos(x)
z = np.sin(x)
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.plot(U, V)
plt.show()
```

```python
array_cols = [10, 20, 30, 40, 50]
array_rows = [1, 2, 3, 4, 5]
plt.xticks(array_cols)
plt.yticks(array_rows)

XX, YY = np.meshgrid(array_cols, array_rows)
plt.plot(XX, YY, marker=".", color='k', linestyle='none')

plt.show()
```

```python

def f(x):
    return 3 * x ** 2 - 4 * x

def df(x):
    return 6 * x - 4

def tf(x):
    return 2 * x - 3


x = np.arange(0, 3, 0.1)
dx = df(x)
tx = tf(x)


plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label="y = 3x^2 - 4x")
plt.plot(x, dx, label="d/dx f(x)")
plt.plot(x, tx, label="Tangent line (x=1)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = 3x^2 - 4x and tangent line at x=1")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the function
def f(x, y):
    return np.log(x**2 + y**2)

    #     return x**2 + y**2

# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of points
x, y = np.meshgrid(x, y)

# Calculate z values
z = f(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
