## 2.4.1. Derivatives


```python
import numpy as np


def f(x):
    return 3 * x ** 2 - 4 * x


for h in 10.0 ** np.arange(-1, -6, -1):
    print(f'h={h:.5f}, numerical limit={(f(1 + h) - f(1)) / h:.5f}')
```

```python
x = np.arange(0, 3, 0.1)
y = f(x)  # calculate the y values using your function
np.column_stack((x, y))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = 3x^2 - 4x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = 3x^2 - 4x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Let's show tangent line at x=1
"""


# Derivative of the function
def tangent(x):
    return 2 * x - 3


x = np.arange(0, 3, 0.1)

plt.figure(figsize=(8, 6))
plt.plot(x, f(x), label="y = 3x^2 - 4x")
plt.plot(x, tangent(x), label="Tangent line (x=1)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = 3x^2 - 4x and tangent line at x=1")
plt.legend()
plt.grid(True)
plt.show()
```

A tangent line to a function at a specific point is a straight line that "just touches" the curve at that point. Mathematically, the tangent line represents the best linear approximation of the function near that point.

If we have a function f(x), the derivative of f at a point a, denoted as f'(a) or df/dx|_(x=a), represents the slope of the tangent line to the function f(x) at the point (a, f(a)).

So, to summarize: The derivative of a function at a certain point gives us the slope of the tangent line to the function at that point. This ties together the geometric idea of a tangent line with the algebraic concept of a derivative.


```python
"""
Let's show tangent line at x=1
"""


def f(x):
    return np.power(x, 3) - 1 / x


def df(x):
    return 2 * x ** 2 + 1 / x ** 2


x = np.arange(0.1, 3, 0.1)
y = f(x)
dy = df(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x)")
plt.plot(x, dy, label="f'(x)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("function and derivative")
plt.legend()
plt.grid(True)
plt.show()
```
