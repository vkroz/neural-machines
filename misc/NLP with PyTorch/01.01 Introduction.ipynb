## Introduction

### TF Representation

TF is a statistical measure of how often a word occurs in a document relative to the total number of words in the document.


```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    'time flies like an arrow',
    'Fruit flies like a banana'
]

one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
```

```python
import seaborn as sns

vocab = one_hot_vectorizer.get_feature_names_out()
sns.heatmap(one_hot, annot=True,
            cbar=False,
            xticklabels=vocab,
            yticklabels=['sentence 1', 'sentence 2'])
```

### TF / IDF Representation

TF-IDF is a statistical measure of how often a word occurs in a document relative to the total number of documents in the corpus.

The IDF representation penalizes words that occur frequently in the corpus.

$$
IDF(w) = log(\frac{N}{df(w)})
$$
where $N$ is the total number of documents in the corpus and $df(w)$ is the number of documents that contain the word $w$.

So, for example, for $N=10$
 - if $df(w)=10$, then $IDF(w)=log(1)=0$.
 - if $df(w)=5$, then $IDF(w)=log(2)=0.301$.
 - if $df(w)=1$, then $IDF(w)=log(10) = 1.0$



TF-IDF score is simply the product of TF and IDF.

$$
TFIDF(w, d) = TF(w, d) \times IDF(w)
$$



```python
print(f'corpus={corpus}')
print(f'vocab={vocab}')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
tfidf
```

```python
sns.heatmap(tfidf, annot=True,
            cbar=False,
            xticklabels=vocab,
            yticklabels=['sentence 1', 'sentence 2'])
```
